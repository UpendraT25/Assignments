```python
a="mY name is UpEndra"
print(f"lower case all letters - {a.lower()}")
print(f"upper case all letters - {a.upper()}")
print(f"capitalize first letter and lower case other - {a.capitalize()}")
print(f"upper case the first letter of each word - {a.title()}")
```
